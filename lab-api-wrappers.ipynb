```python
def get_playlist_tracks(username, playlist_id):

    results = sp.user_playlist_tracks(username, playlist_id)
    tracks = results['items']

    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])

    return tracks

tracks = get_playlist_tracks("spotify", "12Wbv8sIx84T5uh6iOoJ7V")
tracks
```

```python
track_id = [track['track']['uri'] for track in tracks]
len(track_id)
```

```python
# Inicializa una lista para almacenar los artistas y sus IDs
artists_ids = []

# Extrae los artistas y sus IDs
for track in playlist['items']:
    artists = track['track']['artists']
    for artist in artists:
        artist_id = artist['id']
        artist_name = artist['name']
        artists_ids.append({'id': artist_id, 'name': artist_name})
```

```python
import pandas as pd
pd.json_normalize(sp.audio_features(track_id[:100]))
```

```python
start = 0
df = pd.DataFrame()

for stop in range(0, len(track_id),100):
    if start != stop:
        print(start, stop)
        new_df = pd.json_normalize(sp.audio_features(track_id[start:stop]))
        df = pd.concat([new_df, df])
        start = stop

df.reset_index()
```
