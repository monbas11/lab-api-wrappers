```python
def get_playlist_tracks(username, playlist_id):

    results = sp.user_playlist_tracks(username, playlist_id)
    tracks = results['items']

    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])

    return tracks

tracks = get_playlist_tracks("spotify", "12Wbv8sIx84T5uh6iOoJ7V")
tracks
```

```python

# ID de la playlist que quieres analizar
playlist_id = '12Wbv8sIx84T5uh6iOoJ7V'  # Cambia '12Wbv8sIx84T5uh6iOoJ7V' por el ID de tu playlist

# Extraer todas las canciones de la playlist
tracks = []
artists = []
audio_features = []

# FunciÃ³n para extraer las canciones de una pÃ¡gina
def extract_tracks_from_page(page):
    for item in page['items']:
        track = item['track']
        tracks.append(track['name'])
        artist_names = [artist['name'] for artist in track['artists']]
        artists.append(', '.join(artist_names))
        audio_features.append(sp.audio_features(track['id'])[0])

# Obtener la primera pÃ¡gina de la playlist
playlist = sp.playlist_tracks(playlist_id)
extract_tracks_from_page(playlist)

# Mientras haya mÃ¡s pÃ¡ginas, obtener y extraer las canciones
while playlist['next']:
    playlist = sp.next(playlist)
    extract_tracks_from_page(playlist)

# Crear un DataFrame con el nombre de la canciÃ³n y el nombre del artista
df_tracks = pd.DataFrame({'CanciÃ³n': tracks, 'Artista': artists})

# Crear un DataFrame con las caracterÃ­sticas de audio
df_audio_features = pd.DataFrame(audio_features)

# Combinar los DataFrames
df = pd.concat([df_tracks, df_audio_features], axis=1)

# Mostrar el DataFrame
print(df)
```
